# Assignment: Segmenting and Clustering Neighborhoods in Toronto - adding geo

retrieving, cleaning and converting data from wiki

```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)
fnl=pd.DataFrame(dfs[0])
fnl.rename(index=int, columns={0:'PostalCode', 1:'Borough', 2:'Neighborhood'}, inplace=True)
df=fnl[fnl['Borough']!='Not assigned'].copy()
df['Neighborhood'].replace('Not assigned',  df.Borough, inplace=True)
df.drop(0, inplace=True)
fnl= df.groupby('PostalCode').agg({'Borough':'first', 'Neighborhood': ', '.join,}).reset_index()
```

```python
df_geo = pd.read_csv('GeoCord.csv')
df_geo.head()
```

```python
df_geo.rename(index=int, columns={'Postal Code':'PostalCode'}, inplace=True)
```

Merging two DataFrames (adding Latitude and Longitude to main DataFrame)

```python
df_fnl=pd.merge(fnl, df_geo, how='left', on='PostalCode')
```
